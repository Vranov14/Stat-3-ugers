### Task 1: Understanding the NASA API and Data Collection

- Register for a NASA API key and understand the different types of data that the API provides.
- Run the Python script below to fetch data about **Near Earth Objects (NEOs)** from the NASA API for a years data.
- Extract and understand the different pieces of data provided for each NEO.

```python
import requests
import time
from datetime import datetime, timedelta
from getpass import getpass

# Set your NASA API KEY, this step asks you to enter your API KEY.
# (The input box may be float in the top on your editor.)
api_key = getpass()
```

```python
# Set the start and end dates for the data you want to fetch
start_date = datetime.strptime('2022-01-01', '%Y-%m-%d')
end_date = start_date + timedelta(days=365)  # 1 year later

# Initialize a list to store the data
data = []

# Fetch data from the NASA API 7 days at a time
# The introduction of the API is on https://api.nasa.gov, under "Browse APIs" -> "Asteroids NeoWs"
# You can look into the example query in the link below to see what the data look like:
# https://api.nasa.gov/neo/rest/v1/feed?start_date=2015-09-07&end_date=2015-09-08&api_key=DEMO_KEY
current_date = start_date
while current_date < end_date:
    next_date = min(current_date + timedelta(days=7), end_date)
    response = requests.get(f'https://api.nasa.gov/neo/rest/v1/feed?start_date={current_date.strftime("%Y-%m-%d")}&end_date={next_date.strftime("%Y-%m-%d")}&api_key={api_key}')
    data.append(response.json())
    current_date = next_date
    time.sleep(1)  # To avoid hitting the rate limit

# Now 'data' contains the NEO data for the 1-year period
```

```python
# Check the date coverage of your data.
dates_contained_in_data = []
for d in data:
    dates_contained_in_data += list(d['near_earth_objects'].keys())

print(sorted(dates_contained_in_data))
```

### Task 2: Data Analysis

- Calculate the average size of the NEOs for each day.
- Determine the proportion of NEOs that are potentially hazardous.
- Find the NEO with the closest approach distance for each day.
- Use statistical methods to analyze the data. For example, calculate the mean, median, mode, and standard deviation of the NEO sizes. Determine if the size of a NEO is correlated with whether it is potentially hazardous.

```python
from src.utils import get_a_random_chunk_property

# Initialize a dictionary to store the average size for each day
average_sizes = {}

# Iterate over the fetched data
for d in data:
    near_earth_objects = d['near_earth_objects']
    for date_str, neo_list in near_earth_objects.items():
        # Convert date string to datetime object
        date = datetime.strptime(date_str, '%Y-%m-%d')
        
        # Get the sizes of NEOs for the current date
        sizes = [neo['estimated_diameter']['kilometers']['estimated_diameter_min'] for neo in neo_list]
        
        # Calculate the average size
        average_size = sum(sizes) / len(sizes)
        
        # Store the average size for the current date
        average_sizes[date] = average_size

# Print the average sizes for each day
for date, average_size in average_sizes.items():
    print(f"Date: {date}, Average Size: {average_size}")
```

```python
# Initialize a counter for potentially hazardous NEOs
potentially_hazardous_count = 0

# Iterate over the fetched data
for d in data:
    near_earth_objects = d['near_earth_objects']
    for date_str, neo_list in near_earth_objects.items():
        for neo in neo_list:
            # Check if NEO is potentially hazardous
            if neo['is_potentially_hazardous_asteroid']:
                potentially_hazardous_count += 1

# Calculate the total count of NEOs
total_neos_count = sum(len(neo_list) for neo_list in near_earth_objects.values())

# Calculate the proportion of potentially hazardous NEOs
proportion_potentially_hazardous = potentially_hazardous_count / total_neos_count

# Print the proportion
print(f"Proportion of Potentially Hazardous NEOs: {proportion_potentially_hazardous}")
```

```python
# Initialize a dictionary to store the NEO with closest approach distance for each day
closest_neos = {}

# Iterate over the fetched data
for d in data:
    near_earth_objects = d['near_earth_objects']
    for date_str, neo_list in near_earth_objects.items():
        # Convert date string to datetime object
        date = datetime.strptime(date_str, '%Y-%m-%d')

        # Initialize variables for tracking the closest NEO
        closest_neo = None
        closest_distance = float('inf')

        # Iterate over NEOs for the current date
        for neo in neo_list:
            approach_distance = float(neo['close_approach_data'][0]['miss_distance']['kilometers'])

            # Check if current NEO has a closer approach distance
            if approach_distance < closest_distance:
                closest_neo = neo
                closest_distance = approach_distance

        # Store the closest NEO for the current date
        closest_neos[date] = closest_neo

# Print the NEO with closest approach distance for each day
for date, closest_neo in closest_neos.items():
    print(f"Date: {date}, Closest NEO: {closest_neo}")

```

```python
import statistics

# Initialize lists to store NEO sizes and hazardous status
neo_sizes = []
hazardous_status = []

# Iterate over the fetched data
for d in data:
    near_earth_objects = d['near_earth_objects']
    for date_str, neo_list in near_earth_objects.items():
        for neo in neo_list:
            # Extract NEO size and hazardous status
            size = neo['estimated_diameter']['kilometers']['estimated_diameter_min']
            is_hazardous = neo['is_potentially_hazardous_asteroid']
            
            # Add NEO size and hazardous status to respective lists
            neo_sizes.append(size)
            hazardous_status.append(is_hazardous)

# Calculate mean, median, mode, and standard deviation of NEO sizes
mean_size = statistics.mean(neo_sizes)
median_size = statistics.median(neo_sizes)
mode_size = statistics.mode(neo_sizes)
std_dev_size = statistics.stdev(neo_sizes)

# Determine the correlation between NEO size and hazardous status
correlation = statistics.correlation(neo_sizes, hazardous_status)

# Print the calculated statistical measures and correlation
print(f"Mean Size: {mean_size}")
print(f"Median Size: {median_size}")
print(f"Mode Size: {mode_size}")
print(f"Standard Deviation of Size: {std_dev_size}")
print(f"Correlation between Size and Hazardous Status: {correlation}")
```

### Task 3: Data Visualization Part A

- Create a line plot of the number of NEOs per week.
- Create a histogram of the distribution of NEO sizes.
- Create a bar plot of the average NEO size per week.
- Use a library like Seaborn to create more complex visualizations, such as a box plot of the NEO sizes or a heat map of the number of NEOs per week. **Be creative**!

```python
import matplotlib.pyplot as plt

# Initialize lists to store the week numbers and the corresponding number of NEOs
week_numbers = []
neos_per_week = []

# Iterate over the fetched data
for d in data:
    near_earth_objects = d['near_earth_objects']
    for date_str, neo_list in near_earth_objects.items():
        # Convert date string to datetime object
        date = datetime.strptime(date_str, '%Y-%m-%d')
        
        # Extract the week number from the date
        week_number = date.isocalendar()[1]
        
        # Count the number of NEOs for the current week
        num_neos = len(neo_list)
        
        # Append the week number and the number of NEOs to the respective lists
        week_numbers.append(week_number)
        neos_per_week.append(num_neos)

# Create the line plot
plt.plot(week_numbers, neos_per_week)
plt.xlabel('Week Number')
plt.ylabel('Number of NEOs')
plt.title('Number of NEOs per Week')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Initialize a list to store the NEO sizes
neo_sizes = []

# Iterate over the fetched data
for d in data:
    near_earth_objects = d['near_earth_objects']
    for date_str, neo_list in near_earth_objects.items():
        # Extract the NEO sizes for the current date
        sizes = [neo['estimated_diameter']['kilometers']['estimated_diameter_max'] for neo in neo_list]
        
        # Extend the list of NEO sizes
        neo_sizes.extend(sizes)

# Create the histogram
plt.hist(neo_sizes, bins=30)
plt.xlabel('NEO Size (in kilometers)')
plt.ylabel('Frequency')
plt.title('Distribution of NEO Sizes')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

# Create a DataFrame to store the NEO data
neo_data = pd.DataFrame()

# Iterate over the fetched data
for d in data[:500]:  # Limiting the data to 500 entries for a smaller volume
    near_earth_objects = d['near_earth_objects']
    for date_str, neo_list in near_earth_objects.items():
        # Convert date string to datetime object
        date = datetime.strptime(date_str, '%Y-%m-%d')
        
        # Get the estimated diameter of NEOs for the current date
        estimated_diameters = [neo['estimated_diameter']['kilometers']['estimated_diameter_max'] for neo in neo_list]
        
        # Add the NEO estimated diameters to the DataFrame
        neo_data = neo_data.append(pd.DataFrame({'Date': [date] * len(estimated_diameters), 'Estimated Diameter (km)': estimated_diameters}))

# Create a box plot of the NEO estimated diameters
plt.figure(figsize=(10, 6))
sns.boxplot(data=neo_data, x='Date', y='Estimated Diameter (km)')
plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Estimated Diameter (km)')
plt.title('Box Plot of NEO Estimated Diameters')
plt.show()

# Create a heatmap of the number of NEOs per week
neo_counts = neo_data.groupby(neo_data['Date'].dt.week)['Estimated Diameter (km)'].count().reset_index()
neo_counts = neo_counts.pivot(index='Date', columns='Estimated Diameter (km)', values='Estimated Diameter (km)').fillna(0)
sns.heatmap(neo_counts, cmap='YlOrRd', linewidths=0.5)
plt.xlabel('Estimated Diameter (km)')
plt.ylabel('Week')
plt.title('Heatmap of Number of NEOs per Week')
plt.show()
```

```python
import plotly.express as px

# Initialize lists to store data for the scatter plot
neo_names = []
neo_sizes = []
neo_distances = []
neo_hazardous = []

# Iterate over the fetched data
for d in data:
    near_earth_objects = d['near_earth_objects']
    for date_str, neo_list in near_earth_objects.items():
        for neo in neo_list:
            # Extract NEO data
            name = neo['name']
            size = neo['estimated_diameter']['kilometers']['estimated_diameter_max']
            distance = neo['close_approach_data'][0]['miss_distance']['kilometers']
            hazardous = neo['is_potentially_hazardous_asteroid']
            
            # Append NEO data to respective lists
            neo_names.append(name)
            neo_sizes.append(size)
            neo_distances.append(distance)
            neo_hazardous.append(hazardous)

# Create an interactive scatter plot using Plotly
fig = px.scatter(x=neo_distances, y=neo_sizes, hover_name=neo_names, color=neo_hazardous,
                 labels={'x': 'Miss Distance (kilometers)', 'y': 'NEO Size'},
                 title='Interactive Scatter Plot of NEOs')
fig.update_layout(hovermode='closest')
fig.show()
```
